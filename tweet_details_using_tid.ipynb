```python
#tid = df.tweet_id.unique()
for t in tweet_id:
    print("-----------------",t,"-------------")
    try:
        details = api.get_status(t)
        text = details.text
        print("Text: ",text,"\nRetweet count:",details.retweet_count,"\nfavorite_count",details.favorite_count)
        print("Tweet Time: ",details.created_at,"\nFavourited: ",details.favorited,"\nretweeted",details.retweeted,"\nsource: ",details.source)
        
        for key in details.entities:
            extra = details.entities[key]
            size = len(extra)
            if size > 0 and key == 'hashtags': 
                hashtags = [extra[i]['text'] for i in range(0,size)]
                print("Hashtags:",hashtags)
            if size > 0 and (key == 'urls' or key == 'media'):
                k = k + size
                if not urls:   
                    urls = [extra[i]['url'] for i in range(0,size)]
                else:
                    urls.append([extra[i]['url'] for i in range(0,size)])
                print("urls ",urls)
            if size > 0 and key == 'user_mentions':
                mentions = [extra[i]['screen_name'] for i in range(0,size)]
                print("mentions :",mentions)
        print("------------user details---------------")
        print("id",details.user.id,"\nuser.name",details.user.name)
        print("verified:",details.user.verified,"\nfriends_count: ",details.user.friends_count,"\nfollowers_count: ",details.user.followers_count)
        print("listed_count: ",details.user.listed_count,"\nfavourites_count: ",details.user.favourites_count,"\nstatuses_count: ",details.user.statuses_count)

    except tweepy.TweepError as u:
        print(u,t)
```
