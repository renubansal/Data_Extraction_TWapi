```python
import tweepy
import csv
import pandas as pd
from ipynb.fs.full.myapi import *

api = tw_api()

#load tweet details in csv
csvFile = open('tweet.csv', 'a')
csvWriter = csv.writer(csvFile)

#counter of number of tweets
u = 0
for tweet in tweepy.Cursor(api.search,q="#bjp",count=10,lang="en",since="2019-01-01").items():
    u = u + 1
    print(u,tweet.id)
    csvWriter.writerow([tweet.id,tweet.created_at, tweet.text.encode('utf-8')])
```
